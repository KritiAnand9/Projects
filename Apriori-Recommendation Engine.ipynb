### Removing nan values in string to run in Apriori Algorithm

```python
T=[]
for i in range(len(df)):
    T.append([str(df.values[i,j]) for j in range(0,4) if str(df.values[i,j])!='nan'])
```

```python
T
```

```python
pip install apyori
```

```python
from apyori import apriori
```

```python
rules=apriori(T,min_support=0.03,min_confidence=0.35,min_lift=3,min_length=2)
```

```python
list(rules)
```

## Analysis and Interpretation of Results

Result : ordered_statistics=[OrderedStatistic(items_base=frozenset({'JAM'}), items_add=frozenset({'MAGGI'}), confidence=1.0, lift=4.0)

This shows that if a customer purchases base item JAM then there is 100% probability of purchase of MAGGI along with it in one transaction occurrence. We can recommend MAGGI for the customer in section " Products Bought Together " under the product displayed on E-commerce portal.
