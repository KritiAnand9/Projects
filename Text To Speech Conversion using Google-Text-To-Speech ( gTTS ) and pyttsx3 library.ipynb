# Using pyttsx3 library

```python
import pyttsx3
```

```python
engine = pyttsx3.init()
 
engine.say("Speech recognition or text-to-speech conversion may be important in the field of artificial intelligence for people with blindness or deafness or other disabilities.Advances in machine learning make it possible today to convert a speech into text and vice versa with some lines of code.Speech to text conversion means that an app recognizes a words spoken by a person, and converts the voice to written text. ")
engine.setProperty('rate', 80)
engine.setProperty('volume', 0.9)
engine.runAndWait()
```

### We can customise the speed of audio by changing the rate parameter to appropriate level.
